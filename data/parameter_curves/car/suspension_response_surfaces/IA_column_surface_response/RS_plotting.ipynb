```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
RS = pd.read_csv("FL_IA_response_surface.csv")
```

```python
# Lat

fig = plt.figure()
ax = Axes3D(fig, auto_add_to_figure=False)

ax = plt.axes(projection='3d')

fig.add_axes(ax)
ax.scatter(RS["heave"], RS["roll"], RS["IA_change"])
# ax.plot_surface(SR, SA, FY2)
# ax.plot_surface(SR, SA, FY3)

ax.set_xlabel('Heave (m)')
ax.set_ylabel('roll (rad)')
ax.set_zlabel('IA Change (rad)')

ax.invert_xaxis()
```
